### STEP - 1 : Importing Essential Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

### STEP - 2 : Importing the Iris CSV File

```python
pdata = pd.read_csv("Iris.csv")
pdata.head(15)
```

### STEP - 3 : Using describe function

```python
pdata.describe()
```

### STEP - 4 : Collecting information of the given data

```python
pdata.info()
```

### STEP - 5 : Creating the dataframe

```python
import sklearn.datasets as datasets
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))
```

```python
x = pdata.iloc[:, [0, 1, 2, 3]].values
y = pdata["Species"]
dataset_x = pd.DataFrame({'SepalLengthCm': x[:, 0], 'SepalWidthCm': x[:, 1], 'PetalLengthCm': x[:, 2], 'PetalWidthCm': x[:, 3]})
dataset_x.head(10)
```

### STEP - 6 :Creating a target variable 

```python
y= iris.target
dataset_x['target']=y
dataset_x.head()
```

### STEP - 7 : Checking the missing values

```python
dataset_x.apply(lambda i:sum(i.isnull()))
```

### STEP - 8 : Importing the libraries

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import  train_test_split
```

### STEP - 9 : Splitting the data

```python
X = dataset_x.drop('target', axis=1)
y = dataset_x['target']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### STEP - 10 : Applying Decision Tree Algorithm

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

### STEP - 11 : Setting the prediction

```python
pred = dtree.predict(X_test)
```

### STEP - 12 : Visualising the Decision Tree

```python
from sklearn import tree
names = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
classes= ['setosa','versicolour','virginica']
fig,axes= plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(dtree, feature_names=names, class_names=classes,filled=True)
```
